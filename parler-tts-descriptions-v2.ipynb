### Load Dataset

```python
from datasets import load_dataset

dataset = load_dataset("Amadeus99/youtube-transcript-dataset-processed")
```

```python
import random

def generate_description(row):
    templates = [
        "A {gender} delivers a speech in {speech_monotony} manner. The recording is {reverberation} with {sdr_noise}. The pace of speech is {speaking_rate}.",
        "In a {reverberation} environment, a {gender} speaks with {speech_monotony} voice and the speech is {sdr_noise}. The delivery pace is {speaking_rate}.",
        "The speech is given by a {gender} in {speech_monotony} tone. It is recorded in a {reverberation} setting with {sdr_noise}. The speaker delivers it at a {speaking_rate} pace.",
        "A {gender} speaks in a {speech_monotony} tone. The speech is recorded in a {reverberation} environment with {sdr_noise}. The delivery pace is {speaking_rate}.",
        "A {gender} speaker delivers their speech in {speech_monotony} voice. The recording is {reverberation} with {sdr_noise}. The pace of delivery is {speaking_rate}.",
    ]
    
    # Randomly select a template
    template = random.choice(templates)
    
    # Format and return the description
    return template.format(
        gender=row["gender"],
        speech_monotony=row["speech_monotony"],
        reverberation=row["reverberation"],
        sdr_noise=row["sdr_noise"],
        speaking_rate=row["speaking_rate"]
    )
```

```python
processed_dataset = dataset.map(
    lambda row: {"text_description": generate_description(row)},
    batched=False
)
```

```python
processed_dataset.push_to_hub("youtube-transcript-dataset-processed")
```
